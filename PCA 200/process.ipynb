```python
# union final results
tweet_id = [4]
for id in tweet_id:
    p = 'intersection/tweet %d/' % id
    #dirs = os.listdir(p)
    #dirs = ['1 hr', '2 hr', '4 hr']
    dirs = ['6 hr']
    dirs = [p + i + '/' for i in dirs]
    sub = ['first/', 'second/', 'third/']
    for d in dirs:
        col = []
        for s in sub:
            print(d + s)
            for i in range(1, 11):
                df = pd.read_excel(d + s + 'result.xlsx', sheet_name='Batch %d' % i)
                cp  = df.tail(1).copy()
                row = cp.values.flatten().tolist()
                row = row[2:]
                col.append(row)
        df = pd.DataFrame(col)
        df.to_excel(d + 'total_result.xlsx', index= False)
            
```

```python
with open('datagraph_info.json' , 'r') as fp:
    info = json.load(fp= fp)
```

```python
col = []
for datag_id in info.keys():
    row = [datag_id]
    obj = info[datag_id]

    row.append(obj["nodes count"])
    row.append(obj["edge count"])

    train_set = obj["Training Set"]
    row.append(train_set["Positive Samples"])
    row.append(train_set["Negative Samples"])

    test_set = obj["Testing Set"]
    row.append(test_set["Positive Samples"])
    row.append(test_set["Negative Samples"])

    col.append(row)
df = pd.DataFrame(col).T
df
```

```python
df.to_excel('datag_info.xlsx', index=False)
```

```python
col = []
for datag_id in info.keys():
    obj = info[datag_id]
    active = obj["Active/Inactive"]
    row = [datag_id]
    for k in active.keys():
        tmp = active[k]
        row.append(tmp["train active"])
        row.append(tmp["train inactive"])

    for k in active.keys():
        tmp = active[k]
        row.append(tmp["test active"])
        row.append(tmp["test inactive"])
    col.append(row)
act_df = pd.DataFrame(col).T
act_df.to_excel('act_info.xlsx', index=False)
act_df
```
